# Vediamo come se la cava Julia con i cicli for

```julia
rand(Float64)
```

```julia
function test_me()
    n=100
    a=zeros(Float64, n, n, n)
    for i in 1:n, j in 1:n, k in 1:n
        a[i,j,k]=1.
    end
    return a
end
        
```

```julia
test_me()
```

```julia
@benchmark test_me()
```

# Prodotti matriciali

# Integrazione

proviamo a definire la funzione gamma a mano

```julia
#using Pkg
#Pkg.add("QuadGK")
```

```julia
using QuadGK
```

```julia
function gamma_test(z::Int64)
    arg_int(x::Float64) = x^(z-1)*exp(-x)
    quadgk(arg_int,0,Inf)
end
```

```julia
gamma_test(30)
```

```julia
@benchmark gamma_test(30)
```

in confronto alla funzione ufficiale

```julia
@benchmark gamma(30)
```
